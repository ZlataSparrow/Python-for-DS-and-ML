```python
import numpy as np
import pandas as pd

from pandas import Series as s,  DataFrame as df
```

### Concatenating data

```python
obj_df = df(np.arange(36).reshape(6,6))
obj_df
```

```python
obj_df = df(np.random.randint(0,100, size=(4,4)))  # Generates a 4x4 array of random integers between 0 and 100
obj_df
```

```python
obj_2_df = df(np.arange(15).reshape(5,3))
obj_2_df
```

```python
pd.concat([obj_df, obj_2_df], axis=1) #goes and concats further as columns axis = 1
```

```python
pd.concat([obj_df, obj_2_df], axis=0) #goes and concats further as rows axis = 0
```
