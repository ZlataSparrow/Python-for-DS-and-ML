```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame as dfr
```

```python
data = {'name':['steve','john','richard','sarah','randy','micheal','julie'],
        'age':[20,22,20,21,24,23,22],
        'gender':['Male','Male','Male','Female','Male','Male','Female'],
        'rank':[2,1,4,5,3,7,6]}
df=dfr(data)
df
```

### Matplotlib's Bar Chart

```python
plt.bar(df['name'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

```python
plot = sns.barplot(data=df, x='name',y='age')
plot.set_title("Comparing Ages")
plt.show()
```

### Line Plot Matplotlib

```python
plt.plot(df['name'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

### Line Plot Seaborn

```python
plot = sns.lineplot(data=df, x='name',y='age')
plot.set_title("Comparing Ages")
plt.show()
```

### Pie Chart Matplotlib

```python
plt.pie(df['age'], labels = df['name'])
plt.title("Age Comparison")
plt.show()
```

### Pie Chart Seaborn

```python
colors = sns.color_palette('ocean')[0:5]
plt.pie(df['age'], labels = df['name'], colors = colors)
plt.show()
```
