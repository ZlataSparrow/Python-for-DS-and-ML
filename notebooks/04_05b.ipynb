## Creating labels and annotations

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 12,4
sb.set_style('whitegrid')
```

### Labeling plot features
#### The functional method

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]
plt.bar(x,y)
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1,2,3,4,0.5]
veh_type = ['bicycle', 'motorbike','car', 'van', 'stroller']
plt.pie(z, labels =veh_type)
plt.show()
```

#### The object-oriented method

```python
address = '/workspaces/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.5,.5,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation = 60,fontsize = 'medium')
ax.set_title('Miles Per Galon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles per galon')

```

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

#### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.5,.5,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_yticks(range(0, 47, 5))  # Sets y-axis ticks at 0, 5, 10, ..., 45


ax.set_xticklabels(cars.car_names, rotation = 60,fontsize = 'medium')
ax.set_title('Miles Per Galon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles per galon')

ax.legend(loc ='best')
```

### Annotating your plot

```python
mpg.max()
```

```python
fig = plt.figure()
ax = fig.add_axes([.5,.5,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation = 60,fontsize = 'medium')
ax.set_title('Miles Per Galon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles per galon')

ax.set_ylim([0,45])

t_x = cars[cars.car_names == "Toyota Corolla"].index[0]
t_y = cars.loc[cars.car_names == "Toyota Corolla", "mpg"].values[0]
  
ax.annotate('Toyota Corolla', 
            xy=(t_x,t_y), 
            xytext=(21,35),
            fontsize=7,
            weight='bold',
            arrowprops=dict(facecolor='green'))

ax.plot(t_x,t_y, 
        marker='o', 
        color='red', 
        markersize=8, 
        label='Peak')

ax.legend(loc='best')
```
